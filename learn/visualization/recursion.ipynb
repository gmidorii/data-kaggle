# 回帰
- データに直線を当てはめていく

```python
import numpy as np
from numpy.random import randn
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip', tips)
```

```python
sns.lmplot('total_bill','tip', tips,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip', tips, order=5,
          scatter_kws={'marker':'o','color':'indianred'},
          line_kws={'linewidth':1, 'color':'blue'})
```

```python
sns.lmplot('total_bill','tip', tips, fit_reg=False)
```

```python
tips['tips_pect'] = 100*(tips['tip']/tips['total_bill'])
```

```python
tips.head()
```

離散的な値に対しても有効

```python
sns.lmplot('size','tips_pect',tips)
```

x軸に幅をもたせる(ゆらがせる)

```python
sns.lmplot('size','tips_pect',tips, x_jitter=0.2)
```

```python
sns.lmplot('size','tips_pect',tips, x_estimator=np.mean)
```

```python
sns.lmplot('total_bill','tips_pect',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tips_pect',tips,hue='day')
```

# Histogram(ヒストグラム)
- データの分布を可視化する

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
from numpy.random import randn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
ds1 = randn(100)
```

```python
plt.hist(ds1)
```

```python
ds2 = randn(80)
```

```python
plt.hist(ds2, color='indianred')
```

正規化する
- 積分して1になるように描画する

```python
plt.hist(ds1, density=True)
```

グラフを重ね合わせて表示

```python
plt.hist(ds1, density=True, alpha=0.5, bins=20)
plt.hist(ds2, density=True, alpha=0.5, bins=20, color='red')
```

```python
data1 = randn(1000)
data2 = randn(1000)
```

```python
sns.jointplot(data1, data2)
```

```python
sns.jointplot(data1, data2, kind='hex')
```

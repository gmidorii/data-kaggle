# 分布の可視化

````python
from numpy.random import randn
import seaborn as sns
%matplotlib inline
````

````python
ds = randn(100)
````

````python
sns.distplot(ds)
````

````python
sns.distplot(ds, rug=True, hist=False)
````

````python
sns.distplot(ds, bins=25, 
             kde_kws={'color':'indianred','label':'KDE PLOT'},
            hist_kws={'color':'blue', 'label':'HIST PLOT'})
````

````python
from pandas import Series
````

````python
ser1 = Series(ds, name='My_DATA')
````

````python
ser1
````

````python
sns.distplot(ser1)
````
